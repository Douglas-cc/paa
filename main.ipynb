```python
import time as tm
import pandas as pd 
from matplotlib import pyplot as plt
from experimentos import karatsuba, third_grade
```

# Criando Array para os Testes

```python
a_array = [81, 1111, 54541111, 3456123434561234, 34561234345612343456123434561234]
b_array = [32, 1234, 34561234, 5454111134561234, 54541111345612345454111134561234]
```

# Karatsuba

```python
time_karatsuba = []

for indice in range(len(a_array)):
    start = tm.process_time()
    print(karatsuba(a_array[indice], b_array[indice]))
    
    end = tm.process_time()    
    print('Tempo de execução: ', end - start) 
    time_karatsuba.append(end - start)
```

# Third Grade

```python
time_third_grade = []

for indice in range(len(a_array)):
    start = tm.process_time()
    print(third_grade(a_array[indice], b_array[indice]))
    
    end = tm.process_time()    
    print('Tempo de execução: ', end - start) 
    time_third_grade.append(end - start)
```

# Criando DataFrame

```python
n_digitos = [2, 4, 8, 16, 32]
karatsuba = []
third_grade = []

df = pd.DataFrame()
df['n_digitos'] = n_digitos
df['karatsuba'] = time_karatsuba
df['third_grade'] = time_third_grade

df
```

# Grafícos

- Grafíco de Barra

```python
plt.figure(figsize=(10, 6))
plt.style.use('ggplot')

plt.bar(df['n_digitos'], df['third_grade'], width=0.25, label = '3rd Grade', color = 'r')
plt.bar(df['n_digitos'], df['karatsuba'], width=0.25, label = 'Karatsuba', color = 'b')

plt.legend()
plt.title("3rd Grade x Karatsuba")
plt.xlabel('Numero de Digitos')
plt.ylabel('Tempo de Execução')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.style.use('ggplot')

plt.plot(df['n_digitos'], df['third_grade'], label='3rd Grade')
plt.plot(df['n_digitos'], df['karatsuba'], label='karatsuba', linewidth=5)

plt.legend()
plt.xlabel('Numero de Digitos')
plt.ylabel('Tempo de Execução')
plt.title("3rd Grade x Karatsuba")

plt.show()
```
